### Now creating an array containing our dependent variables

```python
x = iris_data.iloc[:, 0:4].values
```

```python
x
```

**Finding the optimum number of clusters using k-means clustering**

Here, I am using the ELbow Method to find the number of optimal number of clusters

```python
# importing kmeans from sklearn's cluster
from sklearn.cluster import KMeans
wcss = []  # wcss- within cluster sum of squares

for i in range(1, 15):
    # "k-means++" is a method to avoid falling into the random initialization trap.
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

**Plotting the the elbow method graph**

```python
plt.figure(figsize=(10, 8))
plt.plot(range(1, 15), wcss, 'r-o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.grid()
plt.show()
```

Here, in the above the graph we can clearly see why it is called as 'The elbow method'.
The optimum clusters is where the elbow occurs. 
And after every iteration the list WCSS dosent decrease with every iteration

In the following graph, we will choose the optimum number of clusters as 3.

**Applying kmeans to the dataset / Creating the kmeans classifier**

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

**Visualising the clusters**

```python
plt.figure(figsize=(10, 8))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
